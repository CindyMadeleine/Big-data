```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
```

```python
#generate data
max = 340 # largest number
min = 100 # smallest number
n = 1000
#Normal distribution 
```

```python
x = np.random.normal(min, max, n)

# Create the bins and histogram
count, bins, ignored = plt.hist(x, 20, normed=True)

# Plot the distribution curve
plt.plot(bins, 1/(max * np.sqrt(2 * np.pi)) *
    np.exp( - (bins - min)**2 / (2 * max**2) ),       linewidth=3, color='y')
plt.show()
```

```python
#generate an array of 10 numbers between min and max 
min = 100 # smallest number
max = 340 # largest number
n = 100
x = np.random.randint(min, max, size=n)
```

```python
fig,ax = plt.subplots(1,1) # return a tuple containing a figure and axes of objects
deg_of_freedom = [4,3,2,1] # df is short for degrees for freedom
linestyles = [':', '--', '-.', '-'] # decide which linestyle to use.

# assign the i'th element in degree_of_freedom to i'th ds
# assign the i'th element in linestyles to ls
# for each element :
#   in the x-axis plot the points of the i'th x
#   in the y-axis plot probability density function given df and x.
#   set the linestyle to the i'th ls.
for df, ls in zip(deg_of_freedom, linestyles):
  ax.plot(x, stats.chi2.pdf(x, df), linestyle=ls) 

#plt.xlim(0, 10) # limit for y-axis
#plt.ylim(0, 0.4) # limit for x-axis

plt.xlabel('Height of trees')
plt.ylabel('Number of trees')
plt.title('Chi-Square Distribution')

#plt.legend()
plt.show()
```
